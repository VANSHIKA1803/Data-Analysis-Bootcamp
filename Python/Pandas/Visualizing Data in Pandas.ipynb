```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Vedant\Desktop\Data Analysis Bootcamp\Python\Pandas\files_samples\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

# LINE Plot

```python
df.plot(kind = 'line')
```

```python
# changing the visuals of graphs

print(plt.style.available)
plt.style.use('seaborn-v0_8-dark-palette')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = "Ice - Cream Ratings", xlabel = "Daily Rating", ylabel = 'Scores')
```

# BAR PLOT

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```
